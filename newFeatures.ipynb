```python
import json
features = json.load(open('features/featDicti.json'))
```

```python
import pandas as pd
training = pd.read_csv('csvs/trainigFiltered.csv')
heldout = pd.read_csv('csvs/heldoutFiltered.csv')
```

```python
for item in features.keys():
    if 'first' in item:
        training[item] = 0
        for i,stuff in enumerate(training['word1']):
            if stuff in features[item]:
                training.ix[i,item] = 1
    if 'second' in item:
        training[item] = 0
        for i,stuff in enumerate(training['word2']):
            if stuff in features[item]:
                training.ix[i,item] = 1
```

```python
training['dvandvaC1More'] = 0
training['word1iu'] = 0
training['dvaca'] = 0
for i,item in enumerate(training['word1']):
    count1 = 0
    count2 = 0
    for stuff in features['Dvandva_ac_list']:
        if stuff in item:
            count1+= 1
        if stuff in training.ix[i,'word2']:
            count2 += 1
        if count1 < count2:
            training.ix[i,'dvandvaC1More'] = 1
            
    if item[-1] == 'i' or item[-1] == 'u':
        training.ix[i,'word1iu'] = 1
    if item[0] in features['Dvandva_ac_list'] and item[-1] == 'a':
        training.ix[i,'dvaca'] = 1
    elif training.ix[i,'word2'][0] in features['Dvandva_ac_list'] and training.ix[i,'word2'][-1] == 'a':
        training.ix[i,'dvaca'] = 1
                                                                                      
```
